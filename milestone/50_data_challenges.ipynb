Although the LendingClub data comes with a data dictionary, the columns named in the data dictionary are not the same as those in the actual data. This necessitated work to determine what some less obviously named features represented.

Another challenge was solved by EDA. The data initially appeared to only contain finalized loans, since it covered data from 2007-2015 and was obtained in 2020. However, the `loan_status` column revealed that almost half of the observations represented ongoing loans. We solved this hurdle by dropping all unfinalized loans when training our model, since their final recovery ratio could not be determined.
