Although the LendingClub data comes with a data dictionary, the columns named in the data dictionary are not the same as those in the actual data. This necessitated work to determine what some less obviously named features represented.

Another challenge was solved by EDA. At first, the data appeared to contain only finalized loans, since it covered data from 2007-2015 and was obtained in 2020. However, the `loan_status` column revealed that almost half of the observations represented ongoing loans at the time the data was compiled. We solved this hurdle by dropping all unfinalized loans when training our model, since their final recovery ratio could not be determined.

Lastly, the choice to use recovery ratio as a target variable was not obvious. We initially considered binary prediction of whether loans were successfully paid off as had been done previously by other researchers [@Change2015]. However, the difference between recovering 10% and 90% of a lendor's initial investment before a debtor defaults is significant; thus, we are choosing to target recovery ratio with our model.
