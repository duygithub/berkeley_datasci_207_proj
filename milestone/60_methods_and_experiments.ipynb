**Experimental Design:** 1. **Data preparation:** Follow Data preprocessing as described before. Apply k-fold cross-validation within training (initially 5 folds) for hyperparameter tuning. Standardise numeric features for linear models to obtain appropriate weights. Tree models are scale-insensitive and do not require scaling. 2. **Hyperparameter tuning**  *Linear models (Ridge / Lasso / Elastic Net):* Overfitting → increase regularization strength (larger α) to reduce variance. Underfitting → decrease regularization or introduce interaction/polynomial terms. *Tree-based models:* Overfitting → reduce max depth, increase regularization, adjust learning rate, apply stronger subsampling, and use early stopping. Underfitting → increase depth or number of trees, reduce regularization, and adjust learning rate. *Neural network models:* Overfitting → increase dropout, add L2 regularization, reduce network size, apply early stopping. Underfitting → increase layers or epochs, reduce regularization, adjust learning rate. 3. **Evaluation metrics:** Mean Absolute Error (MAE) and Root Mean Squared Error (RMSE) to compare predictions against ground truth; lower values indicate better performance. R² for linear models to assess explained variance; higher R² indicates stronger explanatory and forecasting capability. 4. **Comparison method:** Select the model with the lowest MAE and RMSE; linear models should also exhibit high R². Compare train vs. test performance to ensure good generalization. Cross-validation results will also be considered to evaluate robustness.
