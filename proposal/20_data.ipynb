Our data comes from Lending Club, a U.S.-based financial services company which formerly ran a peer-to-peer lending platform that connected individual borrowers with investors. This dataset contains complete loan-level information for all loans originated between the company's founding in 2007 and 2015 [@Forbes2010]. The data includes both borrower characteristics at the time of application and the final loan outcomes, allowing for analysis of credit risk and loan performance over time.

We intend to focus on loan status (e.g., Fully Paid, Current, Late, Charged Off) as a key outcome variable. Informative features include borrower credit quality indicators such as FICO score ranges, debt-to-income ratios, number of past delinquencies, and credit inquiries, as well as loan attributes like interest rate, term length, funded amount, and loan grade. Additional variables capture employment information, income verification status, home ownership, geographic indicators (state and partial ZIP code), and detailed credit history metrics. The dataset consists of approximately 890,000 observations and 75 features, providing a rich foundation for modeling default risk, repayment behavior, and lending outcomes. The data is publicly available through Kaggle: https://www.kaggle.com/datasets/adarshsng/lending-club-loan-data-csv/data [@Kaggle2021].
