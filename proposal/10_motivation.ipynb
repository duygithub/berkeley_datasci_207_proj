During the Great Recession, credit dried up globally. Peer-to-Peer lending platforms like LendingClub and Prosper gain popularity as a way for non-bank actors to provide credit. Due to a confluence of factors including a scandal at LendingClub, the maturing of lending algorithms, operating inefficiencies of P2P lending, and needed scale, LendingClub transitioned away from a P2P model towards institutional lender [@jagtiani_lemieux_2018]. While not directly related to the rise of private credit in the last few years, there are important similarities that can inform the expected ROI. Recently, private credit is becoming an avenue for investors seeking higher-yielding income assets in a diversified portfolio. According to the [New York Federal Reserve @NYFed2025], US household debt continued to increase to 18.59 trillion USD. As investors gain interest in private credit, there are important lessons we can learn from LendingClub's data that will help inform investment decisions in the future when we will inevitably enter another low-liquidity environment.
