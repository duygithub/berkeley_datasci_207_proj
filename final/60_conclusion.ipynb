In this paper, we examined the problem of predicting principal recovery ratios on consumer loans using historical LendingClub data, motivated by the growing importance of private credit markets and the limitations of binary default prediction. Our analysis shows that recovery-based modeling provides meaningful additional insight for investors, particularly given that a nontrivial share of defaulted loans still returns a large fraction of the initial investment. Across models, we observe a clear performance gradient: linear regression establishes a useful but limited baseline, XGBoost captures modest nonlinear improvements, and the LSTM achieves the lowest mean absolute error, indicating a stronger ability to model complex relationships among borrower and loan characteristics. However, this improvement comes with trade-offs, as the LSTM exhibits higher variance and a systematic positive bias in predictions, while XGBoost offers more stable and better-calibrated outputs. These findings suggest that enhanced model structures extract additional signal and improve performance, but one should use caution to interpret results based on their business priority and risk tolerance. Whether average accuracy outweighs systematic bias should be a careful business tradeoff among stakeholders. 

Several limitations of this study should be acknowledged. First, the dataset is restricted to loans issued between 2007 and 2015, which may limit the relevance of the models to current credit environments. Second, despite careful preprocessing, the feature set is limited to variables available at origination and does not incorporate additional data sources such as macroeconomic indicators that could influence recovery outcomes. Third, the use of an LSTM on effectively non-sequential data may not be the most natural architectural choice, potentially contributing to instability and bias. Additionally, the evaluation focuses primarily on standard regression metrics (MAE, RMSE, R²), which do directly translate to investor-relevant considerations such as portfolio-level returns.

There are several avenues for improvement and future work. Incorporating time-varying macroeconomic features—such as unemployment rates or interest rates—could improve model robustness across economic cycles. Experimenting with alternative model architectures may offer improved performance over either the XGBoost or LSTM models. Addressing the bias in predictions through calibration techniques or modified loss functions would also be important for practical deployment. Finally, extending the analysis to a portfolio optimization framework—where predicted recovery ratios inform allocation decisions under different risk preferences—would better connect the modeling results to real-world investment strategies. Overall, this work highlights both the promise and the challenges of moving beyond default prediction toward more nuanced, recovery-focused credit modeling.
