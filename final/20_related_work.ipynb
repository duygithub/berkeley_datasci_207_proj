[@Trinh2024] used a similar peer-to-peer lending data and performed a binary classification (default vs. fully paid) with 9 different statistical and machine learning models. They concluded that Gradient Boosting Decision Tree is the best model for most of the evaluation metrics. [@Change2015] used LendingClub data and fit Logistic Regression, SVM, and Naive Bayes for loan default classification (default vs non-default). They concluded that Naive Bayes with Gaussian performs the best at predicting default rate. Furthermore, by applying the best model, the investment return could grow by 50%. [@Souadda2025] evaluated hyperparameter optimization methods for machine learning models with similar lending data for loan default classification, and Bayesian methods (Hyperopt, Optuna) achieved 75.7 times faster runtime than Grid Search while maintaining similar accuracy.

Existing work focuses on a binary prediction of loan default. Our work steps into a new and more valuable field to predict loan recovery rates based on each loan's characteristics, which is unique among existing literature.
