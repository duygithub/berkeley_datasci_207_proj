[@Trinh2024] used a similar peer-to-peer lending data and performed a binary classification (default vs. fully paid) with 9 different statistical and machine learning models. They concluded that Gradient Boosting Decision Tree is the best model for most of the evaluation metrics. This paper offers a good baseline and an enhanced framework, but it mainly identifies default loans, and it does not predict any economic loss or recovery rates.

[@Change2015] used LendingClub data and fit Logistic Regression, SVM, and Naive Bayes for loan default classification (default vs non-default). They concluded that Naive Bayes with Gaussian performs the best at predicting default rate. Furthermore, by applying the best model, the investment return could grow by 50%. While this paper further provides technical insights into model selection and offers economic intuition at the aggregate level, it does not offer a loan-level prediction and economic interpretation.

[@Souadda2025] evaluated hyperparameter optimization methods for machine learning models with similar lending data for loan default classification, and Bayesian methods (Hyperopt, Optuna) achieved 75.7 times faster runtime than Grid Search while maintaining similar accuracy. However, such a result is only applicable to its original classification setups. While it offers directions for hyperparameter tuning, the exact choice of hyperparameters for our recovery ratio prediction still requires some parameter search and human judgment. 

Overall, existing work focuses on a binary prediction of loan default. Our work steps into a new and more valuable field to predict loan recovery rates based on each loan's characteristics, which is unique among existing literature and valuable to investors.
