Our first model machine learning model used XGBoost. The model operates on tabular feature vectors with no sequential structure.
The model consists of an ensemble of 700 decision trees trained sequentially using gradient boosting.
Each tree has a maximum depth of 4, controlling model complexity and limiting overfitting.
Learning proceeds with a learning rate of 0.04, balancing convergence speed and generalization.
To improve robustness and reduce variance, stochastic subsampling is applied at the row (0.85) and feature (0.9) level.
L2 regularization is used. The objective function is squared error loss.

The XGBoost model was trained using the same feature set and fixed hyperparameters to ensure comparability with the baseline. Stochastic subsampling and L2 regularization controlled for overfitting. This model achieved improved performance, with validation MAE of 0.197 and test MAE of 0.197, along with a test RMSE of 0.275. The corresponding test R² increased to 0.142, indicating that the model captures additional nonlinear structure in the data. However, overall the performance only minorly improves over the linear baseline. Feature importance analysis provides insight into which variables most strongly influenced predictions.
