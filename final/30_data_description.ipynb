The dataset we pulled from Kaggle derives from LendingClub, a U.S.-based firm that operated a peer-to-peer lending platform. The dataset covers all loans originated between 2007 and 2015 and includes borrower information, loan characteristics, and final loan outcomes.

The raw data contains 2,260,668 observations with 145 features [@Kaggle2021]. We are only keeping finalized loans—those with a `"Default"`, `"Charged Off"`, or `"Fully Paid"` status—and the input features `grade`, `sub_grade`, `term`, `emp_length`, and `annual_inc`. We added two input features `issue_year` and `issue_month` derived from the `issue_d` feature. We aim to predict principal recovery ratio, which is the ratio of the features `total_rec_prncp` and `loan_amnt`.

We are splitting the data by time to avoid leakage. The training data consists of 912,546 observations in 8 input features. The validation data, which follows the training data chronologically, consists of 199,546 observations. The test data—last chronologically—consists of 199,546 observations as well.
