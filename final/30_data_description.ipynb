The dataset we pulled from Kaggle derives from LendingClub, a U.S.-based firm that operated a peer-to-peer lending platform. The dataset covers all loans originated between 2007 and 2015 and includes borrower information, loan characteristics, and final loan outcomes.

The raw data contains 2,260,668 observations with 145 features [@Kaggle2021]. We are only keeping finalized loans—those with a `"Default"`, `"Charged Off"`, or `"Fully Paid"` status—and the input features `loan_amnt`, `grade`, `sub_grade`, `term`, `emp_length`, `annual_inc`, `int_rate`, `installment`, `home_ownership`, `dti`, `earliest_cr_line`, `open_acc`, `revol_util`, `verification_status`, `purpose`, `addr_state`, `delinq_2yrs`, `inq_last_6mths`, `pub_rec`, `revol_bal`, `total_acc`, `initial_list_status`, `collections_12_mths_ex_med`, `application_type`, `acc_now_delinq`, `mort_acc`, `pub_rec_bankruptcies`, and `tax_liens`. Categorical and numerical features are processed separately using a preprocessing pipeline. Numerical variables (23 in total) are imputed using the median and standardized, while categorical variables (6 in total) are imputed using the most frequent category and encoded via one-hot encoding. After preprocessing, the feature matrix expands to 101 dimensions. We aim to predict principal recovery ratio, which is the ratio of the features `total_rec_prncp` and `loan_amnt`.

We are splitting the data by time to avoid leakage. The training data consists of 912,546 observations in 8 input features. The validation data, which follows the training data chronologically, consists of 199,546 observations. The test data—last chronologically—consists of 199,546 observations as well.
