The dataset we pulled from Kaggle derives from LendingClub, a U.S.-based firm that operated a peer-to-peer lending platform. The dataset covers all loans originated between 2007 and 2015 and includes borrower information, loan characteristics, and final loan outcomes. The raw data contains 2,260,668 observations with 145 features [@Kaggle2021]. We keep only finalized loans—those with a `"Default"`, `"Charged Off"`, or `"Fully Paid"` status—and use 28 origination-only features, including variables such as `loan_amnt`, `grade`, `term`, `annual_inc`, `int_rate`, `home_ownership`, `revol_util`, `purpose`, `addr_state`, `inq_last_6mths`, `revol_bal`, `total_acc`, `application_type`, `acc_now_delinq`, `pub_rec_bankruptcies`, and `tax_liens`. Categorical and numerical features are processed separately using a preprocessing pipeline. We aim to predict principal recovery ratio, defined as `total_rec_prncp / loan_amnt`. We use a random 70/15/15 train-validation-test split (`random_state=42`), yielding 912,546, 195,545, and 195,547 observations, respectively. After adding calendar features from `issue_d` and applying preprocessing, the feature matrix expands to 101 dimensions.
