Peer-to-Peer lending platforms like LendingClub became popular after the Great Recession when credit tightened. Due to several factors, LendingClub transitioned away from a P2P model towards institutional lenders [@jagtiani_lemieux_2018], similar to those now involved in private credit. There is growing interest in private credit because investors seek diversified, higher-yielding investments and because US household debt is at a record 18.59 trillion [New York Federal Reserve @NYFed2025].

[@Change2015] worked on predicting if a loan will default. We believe there is more utility in predicting the recovery ratio. In our EDA, we can see that approximately 6.8% of loans that defaulted still have a positive return for investors because the interest rate is sufficiently high and the loans defaulted later in the term. We want to learn from LendingClub's data to help inform investment decisions in a future credit-tightening cycle.
