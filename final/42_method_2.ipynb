The second machine learning model we built was a Long Short-Term Memory (LSTM) model. Although LSTMs are typically designed for sequential data,
the model was adapted to loan-level observations by representing each sample as a sequence of length one (i.e., a single timestep with multiple features).
The network starts with an LSTM layer with 32 hidden units, configured to return only the final hidden state. A dropout rate of 0.2 is applied within the LSTM layer for regularization. Then, a fully connected (dense) layer with 16 units and ReLU activation follows. Next comes another dropout layer with a rate of 0.2.
The final dense output layer has a single unit, producing a continuous prediction.
The model is trained using the Adam optimizer and gradient clipping to stabilize training.
The loss function is mean absolute error (MAE).

During training, early stopping and learning rate reduction on plateau are employed based on validation MAE to prevent overfitting and improve convergence.
To capture higher-order interactions among features, a loan-level LSTM model was implemented. While the data are not inherently sequential in time, the LSTM architecture allows for flexible modeling of nonlinear dependencies among input variables. The LSTM achieved the strongest performance in terms of MAE, with a validation MAE of 0.182 and test MAE of 0.178, improving upon the XGBoost model’s test MAE of 0.197. This indicates that the neural network captures additional nonlinear signal present in borrower and loan characteristics. However, the model produced a higher test RMSE of 0.376, reflecting greater sensitivity to large outlier individual errors. A notable limitation is the presence of a positive bias of approximately +0.149 on the test set, indicating systematic overestimation of principal recovery. As a result, while the LSTM improves average accuracy, calibration or loss reweighting would be required to correct this systematic deviation.
