This section presents the results of experiments conducted with the linear regression, gradient boosting (XGBoost), and recurrent neural network (LSTM) models. Experiments across models followed a consistent structure, focusing on feature inclusion using a fixed set of 101 engineered variables, comparative evaluation under fixed hyperparameters, and analysis of model performance across validation and test datasets.

**Baseline Linear Model**:
The linear regression model serves as a baseline using all 101 features without additional feature selection or transformation beyond preprocessing. This model assumes linear relationships between borrower, loan, and contextual variables and the target outcome. The resulting validation MAE was 0.204 with a test MAE of 0.204, and RMSE values of approximately 0.279 on both datasets. The relatively low test R^2 of 0.116 indicates limited explanatory power, suggesting that linear assumptions are insufficient to capture the complexity of the underlying relationships.

**XGBoost Model**:
The XGBoost model was trained using the same feature set and fixed hyperparameters to ensure comparability with the baseline. Stochastic subsampling and L2 regularization controlled for overfitting. This model achieved improved performance, with validation MAE of 0.197 and test MAE of 0.197, along with a test RMSE of 0.275. The corresponding test R^2 increased to 0.142, indicating that the model captures additional nonlinear structure in the data. However, overall the performance only minorly improves over the linear baseline. Feature importance analysis provides insight into which variables most strongly influenced predictions. Feature importance analysis provided insight into which variables most strongly influenced predictions. Credit quality indicators dominated predictive performance, while contextual and geographic factors were less important.

**LSTM Model**:
To capture higher-order interactions among features, a loan-level LSTM model was implemented. While the data are not inherently sequential in time, the LSTM architecture allows for flexible modeling of nonlinear dependencies among input variables. The LSTM achieved the strongest performance in terms of MAE, with a validation MAE of 0.180 and test MAE of 0.178, improving upon the XGBoost model’s test MAE of 0.197. This indicates that the neural network captures additional nonlinear signal present in borrower and loan characteristics. However, the model produced a higher test RMSE of 0.376, reflecting greater sensitivity to large individual errors. A notable limitation is the presence of a positive bias of approximately +0.149 on the test set, indicating systematic overestimation of principal recovery. As a result, while the LSTM improves average accuracy, calibration or loss reweighting would be required to correct this systematic deviation.

**Performance of Final Models**:
The performance of all models demonstrates a clear progression from linear to nonlinear approaches. The XGBoost model consistently outperforms the linear regression baseline across all evaluation metrics, including MAE, RMSE, and R^2. The LSTM further improves MAE, indicating better average predictive accuracy, though this comes at the cost of increased RMSE and prediction bias. Using multiple evaluation metrics provides a more comprehensive understanding of model behavior. MAE captures average error magnitude, while RMSE emphasizes larger deviations. The divergence between these metrics for the LSTM highlights a trade-off between improved average performance and increased variance in prediction errors. To address overfitting, all models were evaluated using a strict train-validation-test split. The similarity between validation and test metrics for both the linear and XGBoost models suggests stable generalization. The LSTM also maintains consistent MAE across splits, though its higher RMSE and bias indicate some degree of overfitting to specific patterns in the training data. Overall, both XGBoost and LSTM models outperformed the linear baseline. The LSTM achieves the lowest average error and demonstrates the strongest ability to model complex relationships, while XGBoost provides more stable and less biased predictions. As such, the choice between models depends on the relative importance of minimizing average error versus maintaining calibration and robustness across predictions.
