This section presents the results of experiments conducted with the linear regression, gradient boosting (XGBoost), and recurrent neural network (LSTM) models. Experiments across models followed a consistent structure, focusing on feature inclusion using a fixed set of 101 engineered variables, comparative evaluation under fixed hyperparameters, and analysis of model performance across validation and test datasets.

Performances are primarily evaluated with MAE (Mean Absolute Error) and RMSE (Root Mean Squared Error). MAE measures the absolute difference between predictions and actual values. RMSE measures prediction error but it squares the errors before averaging, so it penalizes large extreme errors more heavily than MAE. Essentially, a low MAE shows good prediction accuracy on average, and a high RMSE indicates that some extreme values at tail ends likely have larger prediction errors. R^2 is used as a reference metric to evaluate the goodness of fit for the baseline and XGBoost models.

The performance of all models demonstrates a clear progression from linear to nonlinear approaches. The XGBoost model consistently outperforms the linear regression baseline across all evaluation metrics, including MAE, RMSE, and R². The LSTM further improves MAE, indicating better average predictive accuracy, though this comes at the cost of increased RMSE and prediction bias. Using multiple evaluation metrics provides a more comprehensive understanding of model behavior. MAE captures average error magnitude, while RMSE emphasizes larger deviations. The divergence between these metrics for the LSTM highlights a trade-off between improved average performance and increased variance in prediction errors. 

To address overfitting, all models were evaluated using a strict train-validation-test split. The similar MAE and RMSE between validation and test metrics for both the linear and XGBoost models suggest stable generalization. The LSTM also maintains consistent MAE across splits, though its higher RMSE and bias indicate some degree of overfitting to specific patterns in the training data. 

Overall, both XGBoost and LSTM models outperformed the linear baseline. The LSTM achieves the lowest average error and demonstrates the strongest ability to model complex relationships, while XGBoost provides more stable and less biased predictions. As such, the choice between models depends on the relative importance of minimizing average error versus maintaining calibration and robustness across predictions. For a business initiative with lowering the average forecasting error as the priority, LSTM wins as the model choice. For a business environment where risk tolerance is low over prediction errors at extreme values, XGBoost wins as a better model selection.
