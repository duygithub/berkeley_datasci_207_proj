This section presents the results of experiments conducted with the linear regression, gradient boosting (XGBoost), and recurrent neural network (LSTM) models. Experiments across models followed a consistent structure, focusing on feature inclusion using a fixed set of 101 engineered variables, comparative evaluation under fixed hyperparameters, and analysis of model performance across validation and test datasets.


**Performance of Final Models**:

| Model                      | Train MAE | Train RMSE | Val MAE | Val RMSE | Test MAE | Test RMSE | R^2    |
|----------------------------|-----------|------------|---------|----------|----------|-----------|--------|
| Baseline: LinearRegression | 0.2036    | 0.2784     | 0.2041  | 0.2789   | 0.2042   | 0.2790    | 0.1164 |
| XGBRegressor               | 0.1956    | 0.2733     | 0.1968  | 0.2747   | 0.1970   | 0.2750    | 0.1420 |
| LSTM                       | 0.1209    | 0.2883     | 0.1823  | 0.3599   | 0.1780   | 0.3758    |    n/a |

The performance of all models demonstrates a clear progression from linear to nonlinear approaches. The XGBoost model consistently outperforms the linear regression baseline across all evaluation metrics, including MAE, RMSE, and R². The LSTM further improves MAE, indicating better average predictive accuracy, though this comes at the cost of increased RMSE and prediction bias. Using multiple evaluation metrics provides a more comprehensive understanding of model behavior. MAE captures average error magnitude, while RMSE emphasizes larger deviations. The divergence between these metrics for the LSTM highlights a trade-off between improved average performance and increased variance in prediction errors. To address overfitting, all models were evaluated using a strict train-validation-test split. The similarity between validation and test metrics for both the linear and XGBoost models suggests stable generalization. The LSTM also maintains consistent MAE across splits, though its higher RMSE and bias indicate some degree of overfitting to specific patterns in the training data. Overall, both XGBoost and LSTM models outperformed the linear baseline. The LSTM achieves the lowest average error and demonstrates the strongest ability to model complex relationships, while XGBoost provides more stable and less biased predictions. As such, the choice between models depends on the relative importance of minimizing average error versus maintaining calibration and robustness across predictions.
