This study investigates the prediction of recovery ratios for consumer loans to better inform private credit investment strategies, moving beyond traditional binary default classification. Utilizing a LendingClub dataset of finalized loans originated between 2007 and 2015, the methodology compares a baseline linear regression model against two nonlinear machine learning approaches: XGBoost and a loan-level Long Short-Term Memory (LSTM) neural network. Both nonlinear models outperformed the linear baseline. XGBoost provided a modest improvement with stable and well-calibrated outputs, achieving a Test MAE of 0.197. The LSTM model achieved the strongest average predictive accuracy with a Test MAE of 0.178, demonstrating a superior ability to capture complex relationships among borrower and loan characteristics. However, this improvement involved a trade-off, as the LSTM exhibited higher variance (Test RMSE of 0.376) and a systematic positive prediction bias. Ultimately, while recovery-based modeling yields valuable insights for investors, the current models did not achieve the level of accuracy required for practical investment decisions. Despite the progression toward nonlinear methods, the low $R^2$ values suggest that these models require further refinement including the inclusion of macroeconomic indicators.
